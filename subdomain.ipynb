#### PORT SCAN

##### Basic

```python
import nmap

scanner = nmap.PortScanner()

# Define target IP address or hostname
target = "scanme.nmap.org"

# Run a basic scan on the target
scanner.scan(target)

# Print the scan results
for host in scanner.all_hosts():
    print("Host: ", host)
    print("State: ", scanner[host].state())
    for proto in scanner[host].all_protocols():
        print("Protocol: ", proto)
        ports = scanner[host][proto].keys()
        for port in ports:
            print("Port: ", port, "State: ", scanner[host][proto][port]['state'])
```

##### Advanced

```python
import nmap

scanner = nmap.PortScanner()

# Define target IP address or hostname
target = "scanme.nmap.org"

# Define Nmap options
options = "-sS -sV -O -A -p 1-1000"

# Run the Nmap scan with the specified options
scanner.scan(target, arguments=options)

# Print the scan results
for host in scanner.all_hosts():
    print("Host: ", host)
    print("State: ", scanner[host].state())
    for proto in scanner[host].all_protocols():
        print("Protocol: ", proto)
        ports = scanner[host][proto].keys()
        for port in ports:
            print("Port: ", port, "State: ", scanner[host][proto][port]['state'])
```

#### TOOLS

```python
from sublist3r import main
import nmap
import subprocess

def find_subdomain(domain='tesla.com'):

    engines = "baidu,yahoo,google,bing,ask,threatcrowd,ssl,passivedns,netcraft,dnsdumpster,virustotal"

    subdomains = main(domain, 40, None, ports=None, silent=True, verbose=False, enable_bruteforce=False, engines=engines)

    return subdomains

def find_port_details(target="scanme.nmap.org"): # Define target IP address or hostname

    scanner = nmap.PortScanner()

    # Define Nmap options
    # options = "-sS -sV -O -A -p 1-1000"
    # Run the Nmap scan with the specified options
    # scanner.scan(target, arguments=options)

    scanner.scan(target)

    port_details = ""

    for host in scanner.all_hosts():
        port_details += f"Host: {host}\n"
        port_details += f"Sate: {scanner[host].state()}\n"
        for proto in scanner[host].all_protocols():
            port_details += f"Protocol:  {proto}\n"
            ports = scanner[host][proto].keys()
            for port in ports:
                port_details += f"Port: {port} State: {scanner[host][proto][port]['state']}\n"
    
    return port_details

def find_service_fingerprint(link='https://www.example.com'):
    
    process = ['whatweb', f'{link}']

    return subprocess.run(process, capture_output=True, text=True, check=True).stdout

def find_osint_collection(domain="tesla.com"):

    command = ["theharvester", "-d", f"{domain}", "-l", "100", "-b", "bing,yahoo,duckduckgo"]
    result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

    return result.stdout
```
